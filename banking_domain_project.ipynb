```python
bins = [0, 100000, 250000, 600000]

labels = ["Low Income", "Mid Income", "High Income"]

df["income_band"] = pd.cut(
    df["estimated_income"],
    bins=bins,
    labels=labels,
    right=False
)
```

```python
categorical_cols = [
    "location_id",
    "banking_contact",
    "nationality",
    "occupation",
    "fee_structure",
    "loyalty_classification",
    "income_band",
    "brid",
    "genderid",
    "iaid"
]

if "income_band" not in df.columns:
    bins = [0, 100000, 250000, 500000]
    labels = ["Low Income", "Mid Income", "High Income"]
    df["income_band"] = pd.cut(df["estimated_income"], bins=bins, labels=labels)

for col in categorical_cols:
    if col not in df.columns:
        print(f"\nSkipping missing column: {col}")
        continue
    print(f"\nValue counts for {col}:")
    display(df[col].value_counts())
```

```python
df["income_band"].value_counts().plot(kind="bar")
plt.title("Distribution of Income Bands")
```

```python
## Define numerical columns
numerical_cols = [
    "age",
    "estimated_income",
    "superannuation_savings",
    "amount_of_credit_cards",
    "credit_card_balance",
    "bank_loans",
    "bank_deposits",
    "checking_accounts",
    "saving_accounts",
    "foreign_currency_account",
    "business_lending",
    "properties_owned",
    "risk_weighting"
]

```

```python
##summary statistics 
df[numerical_cols].describe()
```

```python
## numerical columns
for col in numerical_cols:
    plt.figure(figsize=(6,4))
    sns.histplot(df[col], kde=True)

    plt.title(f"Distribution of {col}")
    plt.show()
```

```python
## bivariated analysis

# Income vs Bank Deposits
plt.figure(figsize=(6,4))
sns.scatterplot(x=df["estimated_income"], y=df["bank_deposits"]) 
plt.title("Income vs Bank Deposits")
plt.show()

# Average Income by Nationality
avg_income = df.groupby("nationality")["estimated_income"].mean().sort_values()
plt.figure(figsize=(10,5))
avg_income.plot(kind="bar")
plt.title("Average Income by Nationality")
plt.ylabel("Average Income")
plt.show()

# Correlation Heatmap of Numerical Columns
corr_matrix = df[numerical_cols].corr()
plt.figure(figsize=(12,8))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap of Numerical Columns")
plt.show()
```

## insights of EDA
1. Most customers fall in the middle-income category.
2. Estimated income has a positive relationship with superannuation savings, loans, and deposits.
3. Bank deposits and checking accounts show a very strong correlation (0.84).
4. Saving accounts are also strongly related to bank deposits (0.76).
5. Higher-income customers tend to have higher risk weighting.
6. Customers with higher business lending also tend to have higher bank loans.
7. Age has very little impact on banking behavior in this dataset.
8. Property ownership shows almost no correlation with financial variables.
9. Customers with higher deposits generally maintain higher balances across multiple banking products.
10. The dataset indicates strong cross-product banking engagement among financially active customers.
